```python
radec = [];
types = [];
for i in range(len(data)):
    radec.append(data[i].split(' ')[1])
    types.append(data[i].split(' ')[17])
```

```python
ra = []
dec = []
for ii in range(len(radec)):
    ra.append((radec[ii][0:9]))
    dec.append((radec[ii][9:]))
```

```python
def splitra(string):
    h = string[0:2]
    m = string[2:4]
    s = string[4:]
    return h+'h'+m+'m'+s+'s'

def splitdec(string):
    h = string[0:3]
    m = string[3:5]
    s = string[5:]
    return h+'d'+m+'m'+s+'s'
```

```python
count = 0

sel1 = 0
sel2 = 10

n_errors =0
spec_path = 'ks_spec/'
for jj in tqdm(range(len(ra[0:5000]))):

        # print('-'*50)
       # print(count)
        count += 1        #print(name)
        #print('RA, Dec:', obj['ra'], obj['dec'])
        #if count < 3284:
        #    continue
        coorstr = splitra(ra[jj])+' '+splitdec(dec[jj])
        print(coorstr)
        coor = SkyCoord(coorstr, frame='icrs')
        try:
            xid = SDSS.query_region(coor, spectro=True)
        except ValueError:
            print(' Some errors were detected! But I dont know why...')

        if xid == None:
            print('no spec')
            continue
        objid = xid['objid'][0]
        print(xid)
        sp = SDSS.get_spectra(matches=xid)
        print('len(sp): ', len(sp))
        for i in range(len(sp[0:1])):
            try:
                sp[i].writeto(spec_path + str(objid) + '_' + types[jj].replace('/','-') + '_spec_' + str(i) + '.fits')
            except OSError:
                print("File already exists!")
```
