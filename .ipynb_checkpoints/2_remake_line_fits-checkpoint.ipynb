```python
def get_spectrum(source_id):
    path = wdtable[wdtable['source_id'] == source_id][0]['specpath']
    try:
        f = fits.open(path)
        flux = f[1].data['flux']
        loglam = f[1].data['loglam']
        f.close()
        lam = 10**loglam
    except:
        print('Spectrum file missing...')
        return
    return lam,flux
```

```python
wdtable = Table.read('filetable.fits')
```

```python
specs = [];
for i in tqdm(range(len(wdtable[0:100]))):
    spec = get_spectrum(wdtable['source_id'][i])
    specs.append(spec)
```

```python
i = 1

spec = specs[i]
flux = spec[1]
lam = spec[0]
plt.plot(lam,flux)
```

```python
lines = ((lam > hbeta-300) & (lam < hbeta + 300)) | ((lam > halpha-400) & (lam < halpha + 400))\
        | ((lam > hgamma-400) & (lam < hgamma + 200))

contmask = ~lines & (lam > 4050) & (lam < 8000)

skylinemask = (lam > 5578.5 - 10)*(lam < 5578.5 + 10) + (lam > 5894.6 - 10)*(lam < 5894.6 + 10)\
        + (lam > 6301.7 - 10)*(lam < 6301.7 + 10) + (lam > 7246.0 - 10)*(lam < 7246.0 + 10) + \
        (lam > 5173 - 10)*(lam < 5173 + 10)
```

```python
plt.plot(lam,flux)
plt.plot(lam[contmask],flux[contmask],'g.')
```

```python
def quadratic(wl,p1,p2,p3):
    return p1 + p2*wl + p3*wl**2

p,cov = curve_fit(quadratic,lam[contmask],flux[contmask])
```

```python
plt.plot(lam,flux)
plt.plot(lam,quadratic(lam,p[0],p[1],p[2]))
```
