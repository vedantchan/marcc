```python
from astroquery.sdss import SDSS
from astropy.coordinates import SkyCoord
from tqdm import tqdm 
from astropy.table import Table
import astropy.units as u

wdtable = Table.read('wdtable.fits')

count = 0

spec_path = 'spec/'
for obj in tqdm(wdtable[0:1]):
   # print('-'*50)
   # print(count)
    count += 1
    name = str(obj['source_id'])
    #print(name)
    #print('RA, Dec:', obj['ra'], obj['dec'])
    #if count < 3284:
    #    continue
    
    coor = SkyCoord(ra=(obj['ra']-1./3600*0.001*obj['pmra']*15.5)*u.degree, 
                    dec=(obj['dec']-1./3600*0.001*obj['pmdec']*15.5)*u.degree, 
                    frame='icrs')
    
    try:
        xid = SDSS.query_region(coor, spectro=True)
        #print(xid)
    except ValueError:
        print(' Some errors were detected! But I dont know why...')
    
    if xid == None:
        #print('no spec')
        continue
    sp = SDSS.get_spectra(matches=xid)
    #print('len(sp): ', len(sp))
    for i in range(len(sp)):
        try:
            sp[i].writeto(spec_path + name + '_spec_' + str(i) + '.fits')
        except IOError:
            ;
            #print('file already exists')
    
    #break
```
