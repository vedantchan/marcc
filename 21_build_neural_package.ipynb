```python
from wdtools.neural import BaseNN
import numpy as np
import matplotlib.pyplot as plt
```

```python
bnn = BaseNN(n_input = 1, n_output = 1, bayesian = True)
```

```python
mod = bnn.nn()
```

```python
print(mod.summary())
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y)
```

```python
bnn.train(mod, x.reshape(-1,1), y.reshape(-1,1), verbose = 0, n_epochs = 1000)
```

```python
pred = np.ravel(mod.predict(x.reshape(-1,1)))
```

```python
plt.plot(x,y)
plt.plot(x,pred)
```
